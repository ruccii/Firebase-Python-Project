- You can also read data from the database.Using .stream() on the collection() you can able to get the data from the firebase store it inside a variable. Iterate through the variable to get each element in the data

```python
#Read data from Firebase DB ( Firestore )

docs = db.collection(u'Restaurants in Suburbs').stream()

for doc in docs:
    #print(f'{doc.id} => {doc.to_dict()}')
    
    print("ID - ", doc.id)
    
    print("name - ", doc.to_dict().get("name")) #write what is written inside dit
    print("contact - ", doc.to_dict().get("contact")) #write what is written inside dit
    print("address - ", doc.to_dict().get("address")) #write what is written inside dit
    print("imageUrl - ", doc.to_dict().get("imageUrl"))
    print("foodtype - ", doc.to_dict().get("foodtype"))
    print("rest_id - ", doc.to_dict().get("rest_id"))
    
    print("---------------") ##this is for gap
    print("---------------")
```

```python
#Update a Entry -> goto add data -> add data-> update data for code

def updateDatainFirebase(uid, updatedcontact):
    doc_ref = db.collection(u'Restaurants in Suburbs').document(uid)
    
    doc_ref.update({"contact":updatedcontact})
```

- Finally to delete a collection Use delete() with the document(ID) inside the collection.

```python
# Delete Data from Firebase, goto add and manage data-> delete data for code

def deleteDataInFirebase(uid):
    db.collection(u'Restaurants in Suburbs').document(uid).delete()
```
